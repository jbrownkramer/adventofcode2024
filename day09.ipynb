```python
import heapq
```

```python
with open("9.in") as f:
    D = f.read()
lines = D.split("\n")
line = lines[0]
```

```python
def s(a,b):
    return b*(b+1)//2 - (a-1)*a//2
```

```python
rle = list(map(int,line))

#Layout the disk
l = 0
r = len(rle) - (len(rle) % 2) 

checksum = 0
index = 0
while l < r:
    #Try to make initial state always pointing at files
    checksum += s(index,index+rle[l]-1) * (l // 2)
    #print(str(l//2) * rle[l],end="")
    index += rle[l]
    rle[l] = 0
    
    l += 1
    while l < r:
        # Fill up the gap at l
        if rle[r] < rle[l]:
            #print(str(r//2) * rle[r],end="")
            checksum += s(index,index+rle[r]-1)*(r//2)
            index += rle[r]
            rle[l] -= rle[r]
            r -= 2
        else:
            checksum += s(index,index+rle[l]-1)*(r//2)
            #print(str(r//2) * rle[l],end="")
            index += rle[l]
            rle[r] -= rle[l]
            rle[l] = 0
            if rle[r] == 0:
                r -= 2
            break
    l += 1

#Handle any remaining
checksum += s(index,index+rle[r]-1)*r//2
#print(str(r//2) * rle[r],end="")
```

```python
rle = list(map(int,line))

heaps = [[] for i in range(10)]

for i in range(1,len(rle),2):
    heapq.heappush(heaps[rle[i]],i)
    
r = len(rle) - 1 - ((len(rle) - 1) % 2)

index_sum = []
t = 0
for v in rle:
    index_sum.append(t)
    t += v

result = 0
while r >= 0:
    min_i = None
    which_heap = None
    for i in range(rle[r],10):
        if len(heaps[i]) == 0:
            continue
        if min_i is None or heaps[i][0] < min_i:
            min_i = heaps[i][0]
            which_heap = i
        
    if min_i is None or min_i > r:
        result += s(index_sum[r],index_sum[r]+rle[r]-1)*(r // 2)
        r -= 2
        continue
        
    result += s(index_sum[min_i],index_sum[min_i] + rle[r]-1) * (r // 2)
    heapq.heappop(heaps[which_heap])
    heapq.heappush(heaps[which_heap - rle[r]],min_i)
    index_sum[min_i] += rle[r]
        
    r -= 2
        
    
        
    
```
