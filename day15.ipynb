```python
with open("15.in") as f:
    d = f.read()
    
lines = d.split("\n")
```

```python
grid = None
instructions = None

def set_grid():
    global grid
    global instructions
    #Grid
    grid = []
    for i,line in enumerate(lines):
        if len(line)==0:
            break
        row = [c for c in line]
        grid.append(row)

    #Instructions
    instructions = ""
    for line in lines[i:]:
        instructions += line.strip()
        
def get_loc(grid):
    #Loc
    for boti,row in enumerate(grid):
        if "@" in row:
            botj = row.index("@")
            return boti, botj
        
```

```python
dirs = {"<" : (0,-1), ">" : (0,1), "^" : (-1,0), "v" : (1,0)}
m = len(grid)
n = len(grid[0])

def step(grid,boti,botj,instruction):
    di,dj = dirs[instruction]
    
    ni = boti + di
    nj = botj + dj
    
    thing = grid[ni][nj]
    
    if thing == "#":
        return (boti,botj)
    
    if thing == ".":
        grid[boti][botj] = "."
        grid[ni][nj] = "@"
        return ni,nj
    
    thingi = ni
    thingj = nj
    while thing == "O":
        thingi += di
        thingj += dj
        thing = grid[thingi][thingj]
        
    if thing == "#":
        return (boti,botj)
    
    grid[boti][botj] = "."
    grid[ni][nj] = "@"
    grid[thingi][thingj] = "O"
    return (ni,nj)

def display(grid):
    for row in grid:
        for c in row:
            print(c,end="")
        print()
    
```

```python
def fatten(grid):
    to_return = []
    for row in grid:
        new_row = []
        for c in row:
            if c == "#":
                new_row.extend(["#","#"])
            elif c == ".":
                new_row.extend([".","."])
            elif c == "@":
                new_row.extend(["@","."])
            else:
                new_row.extend(["[","]"])
        to_return.append(new_row)
        
    return to_return
```

```python
def step2(grid,boti,botj,instruction):
    di,dj = dirs[instruction]
    
    if di == 0:
        return step_hor(grid,boti,botj,instruction)
    
    return step_vert(grid,boti,botj,instruction)
    
def step_hor(grid,boti,botj,direction):
    di,dj = dirs[direction]
    ni = boti
    nj = botj + dj
    
    thing = grid[ni][nj]
    
    if thing == ".":
        grid[ni][nj] = "@"
        grid[boti][botj] = "."
        return ni,nj
        
    if thing == "#":
        return boti,botj
    
    thingi = ni
    thingj = nj
    
    while thing in ["[","]"]:
        thingj += dj
        thing = grid[thingi][thingj]
        
    if thing == "#":
        return boti,botj
    
    for j in range(thingj,nj-dj,-dj):
        grid[ni][j] = grid[ni][j-dj]
    
    grid[boti][botj] = "."
    return ni,nj

def step_vert(grid,boti,botj,direction):
    di,dj = dirs[direction]
    ni = boti + di
    nj = botj + dj
    
    thing = grid[ni][nj]
    
    if thing == ".":
        grid[ni][nj] = "@"
        grid[boti][botj] = "."
        return ni,nj
    
    if thing == "#":
        return boti,botj
    
    #And now the poop
    pushed = what_is_pushed(grid,boti,botj,di)
    if len(pushed) == 0:
        #Ran into a wall
        return boti,botj
    pushed = sorted(pushed,key=lambda p : -1 * di * p[0])
    for p in pushed:
        grid[p[0]+di][p[1]] = grid[p[0]][p[1]]
        grid[p[0]][p[1]] = "."
    grid[boti][botj] = "."
    grid[boti+di][botj] = "@"
        
    return boti + di,botj

def what_is_pushed(grid,boti,botj,di):
    '''Return the i,j positions of all of the boxes parts that will move as a result of this push'''
    
    to_return = set()
    
    current_row = set([(boti + di,botj)])
    if grid[boti + di][botj] == "[":
        current_row.add((boti + di,botj+1))
    else:
        current_row.add((boti + di,botj-1))
        
    while len(current_row) > 0:
        to_return |= current_row
        next_row = set()
        for ni,nj in current_row:
            thing = grid[ni + di][nj]
            if thing == "#":
                return set()
            if thing in ["[","]"]:
                next_row.add((ni + di,nj))
                if thing == "[":
                    next_row.add((ni + di,nj+1))
                else:
                    next_row.add((ni + di,nj-1))
        current_row = next_row
        
    return to_return
```
