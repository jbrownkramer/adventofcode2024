```python
from functools import cmp_to_key
```

```python
with open("5.in","r") as f:
    lines = f.readlines()
    lines = [l.strip() for l in lines]
```

```python
def ordered(sequence,rules):
    for i,a in enumerate(sequence):
        for b in sequence[i+1:]:
            if (b,a) in rules:
                return False
            
    return True
```

```python
rules = set()
for i,l in enumerate(lines):
    if l=="":
        break
    rules.add(tuple(map(int,l.split("|"))))

t = 0
for l in lines[i+1:]:
    sequence = [int(p) for p in l.split(",")]
    if ordered(sequence,rules):
        t+=sequence[len(sequence)//2]
```

```python
def cmp(a,b):
    return -1 if (a,b) in rules else 1

rules = set()
for i,l in enumerate(lines):
    if l=="":
        break
    rules.add(tuple(map(int,l.split("|"))))

t = 0
for l in lines[i+1:]:
    sequence = [int(p) for p in l.split(",")]
    if not ordered(sequence,rules):
        sequence = sorted(sequence,key=cmp_to_key(cmp))
        t+=sequence[len(sequence)//2]
```
