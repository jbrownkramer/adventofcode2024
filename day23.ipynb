```python
from collections import defaultdict
```

```python
with open("23.in") as f:
    d = f.read()
    lines = d.split("\n")
```

```python
graph = defaultdict(set)
```

```python
for l in lines:
    u,v = l.split("-")
    graph[u].add(v)
    graph[v].add(u)
```

```python
for k in graph:
    graph[k] = sorted(graph[k])
```

```python
t = 0
for v in graph:
    if v[0] != "t":
        continue
    neighbors = graph[v]
    n = len(neighbors)
    for i, u in enumerate(neighbors):
        if u[0] == "t":
            if u < v:
                continue
        for j in range(i + 1, n):
            w = neighbors[j]
            if w[0] == "t" and w < v:
                continue
            if w in graph[u]:
                t += 1
            
```

```python
maximal_cliques = set()

for v in graph:
    if len(maximal_cliques) == 0:
        maximal_cliques.add(v)
        continue
    neighbors = graph[v]
    new_cliques = set()
    new_maximal = set()
    for c in maximal_cliques:
        new_clique = set(c) & neighbors
        new_clique.add(v)
        new_cliques.add(tuple(sorted(new_clique)))
        if len(new_clique) != len(c) + 1:
            new_maximal.add(c)
    maximal_cliques = new_maximal
    new_cliques = sorted(new_cliques,key=lambda c : len(c))
    i = 0
    while i < len(new_cliques):
        c = new_cliques[i]
        for j in range(i+1,len(new_cliques)):
            c2 = new_cliques[j]
            if len(c2) == len(c):
                continue
            if c < c2:
                new_cliques.pop(i)
                break
        else:
            i += 1
    maximal_cliques |= set(new_cliques)
    
```
