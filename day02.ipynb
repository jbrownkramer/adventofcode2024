```python
with open("2.in","r") as f:
    lines = f.readlines()
```

```python
lines
```

```python
def safe(a):
    if len(a) <= 1:
        return True
    
    last = None
    increasing = None
    
    for thing in a:
        if last is None:
            last = thing
            continue
            
        if not (1 <= abs(last - thing) <= 3):
            #print("big gap", last, thing)
            return False
            
        if increasing is None:
            increasing = thing > last
            
        if (thing > last) != increasing:
            #print("wavy", last, thing, increasing)
            return False
        
        last = thing
        
    return True
            
def meta_safe(a):
    for i in range(len(a)):
        if safe(a[:i] + a[i+1:]):
            return True
        
    return False
```

```python
t = 0
for l in lines:
    parts  = [int(p) for p in l.split()]
    if safe(parts):
        t += 1
t
```

```python
t = 0
for l in lines:
    parts  = [int(p) for p in l.split()]
    if meta_safe(parts):
        t += 1
t
```
