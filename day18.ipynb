```python
with open("18.in") as f:
    d = f.read()
```

```python
lines = d.split("\n")
locs = [tuple(map(int,l.split(","))) for l in lines]
```

```python
# n = 7
# nb = 12

n = 71
nb = 1024
```

```python
grid = [[True for j in range(n)] for i in range(n)]
```

```python
for j,i in locs[:nb]:
    grid[i][j] = False
```

```python
for row in grid:
    for free in row:
        if free:
            print(".",end="")
        else:
            print("#",end="")
    print()
```

```python
frontier = {(0,0) : 0}
best = [[None for j in range(n)] for i in range(n)]
ds = [(-1,0),(1,0),(0,-1),(0,1)]

while len(frontier) > 0:
    # Pick the min distance frontier
    mind = None
    mini,minj = None,None
    for (i,j),d in frontier.items():
        if mind is None or d < mind:
            mind = d
            mini,minj = i,j

    #Update structures
    frontier.pop((mini,minj))
    best[mini][minj] = mind

    #Update neighbors
    for di,dj in ds:
        ni = mini + di
        nj = minj + dj

        if ni < 0 or ni >= n or nj < 0 or nj >= n or (not grid[ni][nj]) or best[ni][nj] is not None:
            continue

        if (ni,nj) in frontier:
            frontier[(ni,nj)] = min(frontier[(ni,nj)],mind+1)
        else:
            frontier[(ni,nj)] = mind+1
            
            
```
