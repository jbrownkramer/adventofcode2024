```python
def accessible(locs):
    grid = [[True for j in range(n)] for i in range(n)]
    for j,i in locs:
        grid[i][j] = False
    
    frontier = {(0,0) : 0}
    best = [[None for j in range(n)] for i in range(n)]
    ds = [(-1,0),(1,0),(0,-1),(0,1)]

    while len(frontier) > 0:
        # Pick the min distance frontier
        mind = None
        mini,minj = None,None
        for (i,j),d in frontier.items():
            if mind is None or d < mind:
                mind = d
                mini,minj = i,j

        #Update structures
        frontier.pop((mini,minj))
        best[mini][minj] = mind

        #Update neighbors
        for di,dj in ds:
            ni = mini + di
            nj = minj + dj

            if ni < 0 or ni >= n or nj < 0 or nj >= n or (not grid[ni][nj]) or best[ni][nj] is not None:
                continue

            if (ni,nj) in frontier:
                frontier[(ni,nj)] = min(frontier[(ni,nj)],mind+1)
            else:
                frontier[(ni,nj)] = mind+1
                
    return best[n-1][n-1] is not None

            
```

```python
accessible(locs)
```

```python
lb = 1024
ub = len(locs)

while lb < ub:
    m = (lb + ub)//2
    if accessible(locs[:m]):
        lb = m + 1
    else:
        ub = m
```
